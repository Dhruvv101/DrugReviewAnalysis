## Build a Markov Chain Function

Generating a text using markov chain function(generates a Markov chain from the review corpus, generates text using this Markov chain, and prints the generated text.)
generates text based on existing text

.**analyzing a corpus of text data and building a statistical model of how words transition from one to another.**


1. Markov function:
*     corups.split(): tokenize
*     m_dict = defaultdict(list): empty list to store markov chain\
*      (current_word, next_word): markov chain 
2. generate text function: generate text using above fn


<h2>Markov Chain function</h2>

```python
from collections import defaultdict
import random

def markov_chain(corpus):
    '''Generate a Markov chain from the corpus.'''
    # Tokenize the corpus by word
    words = corpus.split()

    # Initialize a default dictionary to hold all of the words and next words
    m_dict = defaultdict(list)

    # Create a zipped list of all of the word pairs and put them in word: list of next words format
    for current_word, next_word in zip(words[0:-1], words[1:]):
        m_dict[current_word].append(next_word)

    # Convert the default dict back into a dictionary

    m_dict = dict(m_dict)
    return m_dict
corpus = ' '.join(dataf['review'].astype(str))
m_dict = markov_chain(corpus)
print(m_dict)
```

# Create Text Generator

<h2>Generating text from markov function</h2>

generate_sentence, takes a Markov chain dictionary markov_dict as input along with the number of words num_words

```python
import random

def generate_sentence(markov_dict, num_words=10, seed=None):
    '''Generate a sentence using the Markov chain dictionary.'''
    # Seed the random number generator for reproducibility
    random.seed(seed)
    
    # Choose a random starting word
    current_word = random.choice(list(markov_dict.keys()))
    
    # Initialize the sentence
    sentence = [current_word]
    
    # Generate the sentence until reaching the desired number of words
    for _ in range(num_words-1):
        # Choose the next word randomly from the list of next words
        next_word = random.choice(markov_dict[current_word])
        
        # Append the next word to the sentence
        sentence.append(next_word)
        
        # Update the current word for the next iteration
        current_word = next_word
    
    # Capitalize the first word of the sentence
    sentence[0] = sentence[0].capitalize()
    
    # Add a period at the end of the sentence
    sentence[-1] += '.'
    
    # Join the words into a single sentence
    sentence = ' '.join(sentence)
    return sentence

# Example usage:
# Example usage:
generated_sentence = generate_sentence(m_dict, num_words=50, seed=123)
print(generated_sentence)

```

```python
import random
import string

def generate_sentence_with_punctuation(chain, max_length=15):
    '''Generate a sentence with the given Markov chain, ending with a random punctuation mark.'''
    word1 = random.choice(list(chain.keys()))
    sentence = [word1.capitalize()]

    while len(sentence) < max_length:
        word2 = random.choice(chain.get(word1, ['']))
        if not word2:
            break
        sentence.append(word2)
        if word2[-1] in string.punctuation:
            break
        word1 = word2

    # Add a random punctuation mark if the sentence doesn't end with one
    if sentence[-1][-1] not in string.punctuation:
        sentence[-1] += random.choice(string.punctuation)

    return ' '.join(sentence)
```
